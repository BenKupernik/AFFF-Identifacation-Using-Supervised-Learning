```python
import pandas as pd
```

This data comes from an enviromental site assessment for an army base in australia. It represents PFAS generated by an AFFF source. Likely from fire fighting training. 
http://www.defence.gov.au/environment/pfas/oakey/

```python
# read in the file, the file has '-' to represent a test for an analyte that wasn't run. 
# na_values sets these values to na when pandas reads it in
df_a = pd.read_excel('Pages from aac oakey.xlsx', na_values = '-')
# drop any row or column that dosn't have any data in it.
df_a = df_a.dropna(axis=1, how='all')
df_a = df_a.dropna(axis=0, how='all')
```

When a lab runs an test for a particular chemical they calculate a limit called the detection limit 
where if the chemical is present in contractions equal to or greater than that limit they will detect it. If they don't detect it, they say it's less than the detect limit because they don’t know if there is any of that chemical below that limit.  In our case we want to set those values to zero so we can compare the data.

```python
for col in df_a.columns:
    df_a[col] = df_a[col].apply(lambda val: 0 if '<' in str(val) else val)
df_a
```

```python
# read in the second table from this source and clean it
df_a_2 = pd.read_excel('Pages from aac oakey.xlsx', na_values = '-', sheet_name = 'Sheet 1')
df_a_2 = df_a.dropna(axis=1, how='all')
df_a_2 = df_a.dropna(axis=0, how='all')
```

```python
# set non detect values to 0 again
for col in df_a_2.columns:
    df_a_2[col] = df_a_2[col].apply(lambda val: 0 if '<' in str(val) else val)
df_a_2
```

```python
# concatanate the two dfs together
df = pd.concat([df_a, df_a_2]).reset_index(drop = True)
```

```python
# fix some names and add a column to say this is an AFFF source
df = df.rename(columns = {'\\PFBS':'PFBS'})
df['AFFF'] = 1
df.columns
```

This data comes from 18 landfills across the US.
'National Estimate of Perand Polyfluoroalkyl Substance (PFAS) Release to US Municipal Landfill Leachate'

```python
# read in some data from a non AFFF source 
df_na = pd.read_excel('landfill_pfas.xlsx')
df_na = df_na.dropna(axis=1, how='all')
df_na = df_na.dropna(axis=0, how='all')
```

```python
df_na
```

```python
# clean up the non detects and other values
for col in df_na.columns:
    df_na[col] = df_na[col].apply(lambda val: 0 if 'ND' in str(val) else val)
    df_na[col] = df_na[col].apply(lambda val: 0 if '<LOQ' in str(val) else val)
    # this data in ng/l (10^-9 g/l) where as the other data is in ug/l (10^-6 g/l) so I need to convert it
    df_na[col] = df_na[col].apply(lambda val: val * 0.001 if isinstance(val, float) else val)
```

```python
# set the afff column to false
df_na['AFFF'] = 0
df_na
```

```python
# now to standerdize the column names
print(df_na.columns)
df.columns
```

```python
# we only need the columns that appeare in both sources and the names need to be the same
df_na = df_na.rename(columns={'4:2 FtS':'4:2 FTS', '8:2 FtS':'8:2 FTS', '6:2 FtS':'6:2 FTS', 'MeFOSAA':'MeFoSAA',
                             })
df_na = df_na[['PFOS', 'PFHxS', 'PFOA', 'PFHxA', '4:2 FTS', '8:2 FTS', '6:2 FTS', 'EtFOSAA', 'MeFoSAA', 'PFBA',
              'PFDS', 'PFHpS', 'PFPeA', 'PFBS', 'PFDA', 'PFDoDA', 'PFHpA', 'PFNA', 'PFTeDA', 'PFUnDA', 'AFFF']]
df = df.drop(columns = ['10:2 FTS', 'EtFOSA', 'EtFOSE', 'MeFOSA', 'MeFOSE', 'PFPeS', 'FOSA', 'PFTrDA', 'Sum of PFAS'])
```

```python
df_na
```

```python
# add both dataframes together
df = pd.concat([df, df_na]).reset_index(drop = True)
```

```python
# normalise the data so if one site has much higher concentrations that won't be used as a trival way to 
# differencate AFFF vs non AFFF
df_afff = df.AFFF
df_rest = df.drop(columns=['AFFF'])
df_rest = df_rest.div(df_rest.sum(axis=1), axis=0)
df_rest['AFFF'] = df_afff
df = df_rest
```

```python
# and export to a csv
df.to_csv('pfas_data.csv', index = False)
```
