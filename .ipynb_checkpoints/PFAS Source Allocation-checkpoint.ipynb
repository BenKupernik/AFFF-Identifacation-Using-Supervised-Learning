```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, f1_score
from sklearn.base import clone 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

Definitions

AFFF - aqueous film forming foam its used as a fire extinguisher for oily fiers


PFAS - Per- and Polyfluorinated Substances. A family of chemicals found in a wide variety of
	   Products. If you have heard of ‘forever chemicals’ that’s what these are.


Inspiration

The idea for this project came from this paper.
https://www.sciencedirect.com/science/article/pii/S0045653520307864
To put this in context let's say you have city A which has a landfill and a military base nearby. 
In the last decade or two doctors have discovered that PFAS are unhealthy and the Environmental Protection Agency (among others) has started to regulate how much PFAS can be in the drinking water for city A. City A then diligently tests its drinking water and finds that there are more PFAS in the water than is allowable under the new regulation. City A now has to build or modify a water treatment facility to remove the PFAS from their water. That's expensive, who’s going to pay for that? Some initial testing from City A shows that both the nearby landfill and military base are sources of PFAS. City A then files a lawsuit with the owners of the landfill and the military base as potentially responsible parties and they now need to figure out who contributed how much contamination, and subsequently how much each of them has to pay.

That's where this project comes in. The PFAS chemical family has hundreds of individual chemicals in it (PFOS, PFHxS, PFOA, PFHxA, 4:2 FTS, 8:2 FTS ….). The ratio of those chemicals can be used to determine the source of the contamination and used to determine how much the landfill or military base contributed. What the authors of the parer did and what I’m doing is seeing if you can use supervised binary classification to determine the source of PFAS contamination.


```python
# read in the data
df = pd.read_csv('pfas_data.csv')
print(df)
```

```python
# In total I have 137 samples and 20 features. 
# That's not enough data for that many features so I need to narrow it down

for col in df.columns:
    # drop any rows where theres NAN values for this column
    df_temp = df.dropna(subset=[col])
    # split the df in to AFFF and Non AFFF
    df_afff = df_temp.drop(df_temp[df_temp.AFFF == True].index)
    df_non_afff = df_temp.drop(df_temp[df_temp.AFFF == False].index)
    # print out how many rows have a non zero value in this column for AFFF/non AFFF
    print('%s has %s for AFFF and %s for non AFFF' % (col, np.count_nonzero(df_afff[col]), np.count_nonzero(df_non_afff[col])))
```

```python
# I picked 4 columns that looked like that have a good mix of nonzero values for AFFF and non AFFF
top_four_analytes = ['PFOS', 'PFHxS', 'PFOA', 'PFBS']
# drop any row where one of these analytes is a NAN
df_top_four = df.dropna(subset=top_four_analytes)
df_top_four
```

```python
# split the data in to train and test sets
x = df_top_four[top_four_analytes]
y = df_top_four.AFFF
X_train, X_test, y_train, y_test = train_test_split(x.to_numpy(), y.to_numpy(), test_size=.5, random_state = 5)
```

```python
# I'd like to try out the KNN method on this data. 
# I'll loop through difrent values of neighbors and see how well they do

# initalize some lists 
acc_train = []
acc_test = []
f1_train = []
f1_test = []

k_range = range(1, 11)

# loop through however many neighbors
for k in k_range:
    # train the model
    knn = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train)
    y_hat_train = knn.predict(X_train)
    y_hat_test = knn.predict(X_test)
    
    # save performance metrics
    acc_train.append(accuracy_score(y_train, y_hat_train))
    acc_test.append(accuracy_score(y_test, y_hat_test))
    f1_train.append(f1_score(y_train, y_hat_train))
    f1_test.append(f1_score(y_test, y_hat_test))
    
# graph it!
fig,(ax1, ax2) = plt.subplots(1, 2, figsize=(12,7), sharey=True)# plt.subplots(nrows=1,ncols=1,figsize=(12,7))
ax1.plot(k_range, acc_train, marker="o", color="steelblue", lw=3, label="Train Accuracy")
ax1.plot(k_range, acc_test, marker="o", color="red", lw=3, label="Test Accuracy")
ax2.plot(k_range, f1_train, marker="o", color="steelblue", lw=3, label="Train F1", linestyle='dashed')
ax2.plot(k_range, f1_test, marker="o", color="red", lw=3, label="Test F1", linestyle='dashed')
ax1.set_xlabel("Number of Neighbors", fontsize=16)
ax1.set_ylabel("Accuracy", fontsize=16)
ax1.legend()
ax1.grid(alpha=0.25)
ax2.set_xlabel("Number of Neighbors", fontsize=16)
ax2.set_ylabel("F1 Score", fontsize=16)
ax2.legend()
ax2.grid(alpha=0.25)
```

KNN Conclusions

Based on both the accuracy score and the f1 score it looks like I can predict whether a sample was generated by AFFF or not with ~95% accuracy. That's not too bad. It looks like K=2 is the best choice. It’s comparable to K=1 but I suspect that might be low of a choice. For comparison the paper found values between 92% and 97% accuracy on their test data for KNN. They found that K = 3 was the best choice. 


```python
# I'm also intrested to see how well a descision tree method would preform. 
# The authors used an extra random forest but I'd like to try ada boosting

# same idea as above, loop through however many learners and graph some metrics

# initalize some lists 
acc_train = []
acc_test = []
f1_train = []
f1_test = []

n_estimators = range(1, 10)

# loop through my learners
for n in n_estimators:
    # fit and train the modle
    clf = AdaBoostClassifier(n_estimators=n, random_state=0)
    adaBoost = clf.fit(X_train, y_train)
    
    # make some predictions
    y_hat_train = adaBoost.predict(X_train)
    y_hat_test = adaBoost.predict(X_test) 
    
    # save the performance metrics
    acc_train.append(accuracy_score(y_train, y_hat_train))
    acc_test.append(accuracy_score(y_test, y_hat_test))
    f1_train.append(f1_score(y_train, y_hat_train))
    f1_test.append(f1_score(y_test, y_hat_test))

# graph it!
fig,(ax1, ax2) = plt.subplots(1, 2, figsize=(12,7), sharey=True)# plt.subplots(nrows=1,ncols=1,figsize=(12,7))
ax1.plot(n_estimators, acc_train, marker="o", color="steelblue", lw=3, label="Train Accuracy")
ax1.plot(n_estimators, acc_test, marker="o", color="red", lw=3, label="Test Accuracy")
ax2.plot(n_estimators, f1_train, marker="o", color="steelblue", lw=3, label="Train F1", linestyle='dashed')
ax2.plot(n_estimators, f1_test, marker="o", color="red", lw=3, label="Test F1", linestyle='dashed')
ax1.set_xlabel("Number of Learners", fontsize=16)
ax1.set_ylabel("Accuracy", fontsize=16)
ax1.legend()
ax1.grid(alpha=0.25)
ax2.set_xlabel("Number of Learners", fontsize=16)
ax2.set_ylabel("F1 Score", fontsize=16)
ax2.legend()
ax2.grid(alpha=0.25)
```

Ada Boost Conclusions

At three learners I can determine AFFF vs non AFFF sources with 98% accuracy and a 96% f1 score. That's not bad. Given that it only takes three learners and I am using stumps that indicates there are some easily identifiable characteristics in the chemistry that are distinct for AFF sources. Since just one learner can get you to 77% accuracy there is definitely a large feature that the algorithm identified. After that the improvements are less drastic. The paper did not have an ada boost method to compare to but their extra trees method was 95% accurate.
